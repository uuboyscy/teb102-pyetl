```python
import jieba
import glob

jieba.load_userdict('./mydict.txt')

fileList = glob.glob('res_gossiping/*.txt')

sb = ''
for file in fileList:
    with open(file, 'r', encoding='utf-8') as f:
        tmpStr = f.read().split('---split---')[0]
    sb += tmpStr
```

```python
sb
```

```python
sb_cut = jieba.cut(sb)
sb_list = [w for w in sb_cut]
sb_list
```

```python
with open('./stopword.txt', 'r', encoding='utf-8-sig') as f:
    tmpList = [i.replace('\n', '') for i in f.readlines()]
tmpList
```

```python

wc_dict = dict()
for w in sb_list:
    if (len(w) > 1) and (w not in tmpList):
        if w in wc_dict:
            wc_dict[w] += 1
        else:
            wc_dict[w] = 1
wc_dict
```

```python
wc_list = [(i, wc_dict[i]) for i in wc_dict]
wc_list
```

```python
wc_list.sort(key=lambda x: x[1], reverse=True)
```
