```python
import requests
import os
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.125 Safari/537.36'
}

os.mkdir('./pttArticle')
```

```python
url = 'https://www.ptt.cc/bbs/movie/index.html'

for i in range(0, 5):
    res = requests.get(url=url, headers=headers) # get response -> <Response [200]>

    html = res.text # html string

    soup = BeautifulSoup(html, 'html.parser')

    title_list = soup.select('div.title')

    for title_soup in title_list:
    #     print(title_soup)
    #     print('===========')
        title_tmp_soup = title_soup.select('a')
#         print(title_tmp_soup)

        if len(title_tmp_soup) == 0:
            pass
        else:
            title = title_tmp_soup[0].text
            title_url = 'https://www.ptt.cc' + title_tmp_soup[0]['href']
            ''
            # Get article content
            res_article = requests.get(title_url, headers=headers)
            soup_article = BeautifulSoup(res_article.text, 'html.parser')
            soup_article_content = soup_article.select('div[id="main-content"]')[0]
            article_content = soup_article_content.text.split('※ 發信站:')[0]
            print(article_content)
            try:
                with open('./pttArticle/{}.txt'.format(title), 'w', encoding='utf-8') as f:
                    f.write(article_content)
            except FileNotFoundError:
                with open('./pttArticle/{}.txt'.format(title.replace('/', '')), 'w', encoding='utf-8') as f:
                    f.write(article_content)
            except OSError:
                pass
            
            print(title)
            print(title_url)

#         print('========')

    # Get last page url
    last_page_url = 'https://www.ptt.cc' + soup.select('a[class="btn wide"]')[1]['href']
#     print(last_page_url)
    url = last_page_url
    
    print('=============')
```
