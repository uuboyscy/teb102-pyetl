```python
import pandas as pd
import os
import glob
```

```python
os.listdir('res_gossiping')
```

```python
glob.glob('res_gossiping/*.txt')
```

```python
fileList = glob.glob('res_gossiping/*.txt')

data = list()
for file in fileList:
    with open(file, 'r', encoding='utf-8') as f:
        tmpStr = f.read().split('---split---')[1]
#     print(tmpStr)
    tmpList = tmpStr.split('\n')[1:-1]
    data.append(tmpList)
#     print(tmpList)
#     print('========')

data
```

```python
columns = [i.split(': ')[0] for i in data[0]]
columns
```

```python
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def column_filter(x):
    output = x.split(': ')[1]
    return output
```

```python
df['推'].apply(column_filter)
```

```python
df
```

```python
df2 = df.copy()
```

```python
df2['推'] = df2['推'].apply(column_filter)
df2
```

```python
for c in columns[1:]:
    df2[c] = df2[c].apply(column_filter)
df2
```

```python
for c in columns:
    df[c] = df[c].apply(lambda x: x.split(': ')[1])
df
```
