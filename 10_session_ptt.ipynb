# Program

```python
import requests
from bs4 import BeautifulSoup

url_landing = 'https://www.ptt.cc/ask/over18?from=%2Fbbs%2FGossiping%2Findex.html'
url_over18 = 'https://www.ptt.cc/ask/over18'

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.125 Safari/537.36'
}

ss = requests.session()
res = ss.get(url_landing, headers=headers)

soup = BeautifulSoup(res.text, 'html.parser')

tmp_key = soup.select('input')[0]['name']
tmp_value = soup.select('input')[0]['value']

yes_key = soup.select('button')[0]['name']
yes_value = soup.select('button')[0]['value']

data = {
    tmp_key: tmp_value,
    yes_key: yes_value
}

ss.post(url_over18, headers=headers, data=data)

res = ss.get('https://www.ptt.cc/bbs/Gossiping/index.html')
print(res.text)
```
